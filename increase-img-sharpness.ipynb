    a)Read the image named lena.tif and increase the sharpness by adding its first derivative estimation, obtained by Sobel operator. What have you obtained?
    
    b)Try to increase the sharpness of lena.tif by adding the first derivative (gradient) and the second derivative (Laplace operator). Which of these methods gives better result?


```python
import cv2
from matplotlib import pyplot as plt
from scipy import ndimage
```

```python
img = cv2.imread('lena.tif')
plt.imshow(img)
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
f, ax = plt.subplots(2,2,figsize=(15,15))
ax[0][0].imshow(img,cmap = 'gray')
ax[0][0].set_title('Original')
ax[0][1].imshow(laplacian,cmap = 'gray')
ax[0][1].set_title('Laplacian')
ax[1][0].imshow(sobelx,cmap = 'gray')
ax[1][0].set_title('Sobel X')
ax[1][1].imshow(sobely,cmap = 'gray')
ax[1][1].set_title('Sobel Y')
plt.show()
```
