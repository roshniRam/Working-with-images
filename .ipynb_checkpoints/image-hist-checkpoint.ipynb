Q1. Read an image named pout.tif, plot its histogram, normalize the histogram and per- form histogram equalization. Comment about the result. Note: Histogram equal- ization is an opera- tion that changes the histogram of an image so that the number of pixels is approximately equal for all brightness levels.

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.image as mpimg
```

```python
img = cv2.imread('pout.tif',-1)
```

```python
plt.hist(img.ravel())
plt.show()
```

# Normalizing the plot

```python
img.shape
```

```python
imgSize = img.shape[0] * img.shape[1] * img.shape[2]
equ = img / img.max()
```

```python
plt.hist(equ.ravel())
plt.show()
```

```python
fig, ax = plt.subplots(1,2)
ax[0].imshow(img)
ax[1].imshow(equ)
```

Histogram equalization is an operation that changes the histogram of an image so that the number of pixels is approximately equal for all brightness levels. As we can see from the plot, the new histogram is more distributed i.e. the overall shape of the histogram has changed and the data is more ditributed.
